# Time Series

In this brief notebook we plot the time series of aggregated temperature excursions from 68F (proxy for heating demand) over the Texas Interconnect.

```python
import matplotlib.pyplot as plt
import pandas as pd
import xarray as xr

from codebase.fig import smart_save
```

Load in the data

```python
hdd = xr.open_dataset("../data/processed/era5/tx/hdd_ercot.nc")
```

Create the plot

```python
plt.figure(figsize=(9, 5))

hdd["pop_weighted"].sel(pop_year=2020, time=slice("2021-02-07", "2021-02-21")).plot(
    label="Population Weighted"
)
plt.xlabel("Time [UTC]")
plt.ylabel("Pop. Weighted Temperature Excursion")
plt.title("")

smart_save("HDD_pop_weighted_ts")
plt.show()
```
